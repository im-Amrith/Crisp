```python
x = v.drop(columns='Disease') # features
y = v["Disease"] # targets
```

```python
x,y
```

```python
import sklearn
print(sklearn.__version__)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
# from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline # A  pipeline streamlines machine learning workflows by chaining together multiple data processing steps and a final estimator (model) into a single unit. This structure simplifies code, enhances readability, and prevents data leakage.
```

```python
ohe = OneHotEncoder() # made a object named ohe of onehotencoder
ohe.fit(x[['Animal','Symptom 1','Symptom 2','Symptom 3']]) # in ohe we will pass categorical values
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown = 'ignore'),['Animal','Symptom 1','Symptom 2','Symptom 3'])
                                         ,remainder = 'passthrough')
```

```python
# lr = LogisticRegression()
# pipe = make_pipeline(column_trans,lr)
rf = RandomForestClassifier(random_state=42)  # You can tweak hyperparameters here
pipe = make_pipeline(column_trans, rf)
# we give raw data to pipeline and it apply tranformer operation and also apply logistic reg to predict target
#Shown above
# if we use ohe.categories and if we provide wrong categories in x_test then it will show  show error
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
import pickle
pickle.dump(pipe,open('livestock_logistic_Model.pkl','wb'))
```
